```python
import os
from pymongo import MongoClient
import pandas as pd
```

```python
REPUTATION_MONGODB_URI = os.getenv("REPUTATION_MONGODB_URI")
```

```python
client = MongoClient(REPUTATION_MONGODB_URI)
reputation_db = client.get_database("reputation")

for collection_name in reputation_db.list_collection_names():
    print(collection_name)
    pd.DataFrame.from_records(reputation_db[collection_name].find()).to_csv(
        f"../data/{collection_name}.csv", index=False
    )
    print(f"Saved {collection_name}.csv")
```
